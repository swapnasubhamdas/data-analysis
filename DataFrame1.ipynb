```python
# Rank Values with the .rank() Method
# We’ll practice this method on a numeric Series and then confirm the results through our own sort test
```

```python
import pandas as pd
nba = pd.read_csv('players.csv').dropna(how= "all")
nba["Player"].fillna(value = "Mising Player Name", inplace = True)
nba["height"] = nba["height"].fillna(value = 175).astype("int")
nba["weight"] = nba["weight"].fillna(value = 70).astype("int")
nba["collage"].fillna(value = "No Collage", inplace = True)
nba["born"] = nba["born"].fillna(value = 1900).astype("int")
nba["birth_city"].fillna(value = "Birth City Missing", inplace = True)
nba["birth_state"].fillna(value = "Birth State Missing", inplace = True)
nba.head()
```

```python
nba["height_rank"] = nba["height"].rank(ascending=False,method = "dense").astype("int")
```

```python
nba.head()
```

```python
nba.sort_values(by="height",ascending= False).head(21)
```

```python
nba["height_rank"] = nba["height"].rank(ascending=False,method = "min").astype("int")
nba.head()
```

```python
nba.sort_values(by="height",ascending= False).head(21)
```
