```python
#The .apply() Method
google_stock_price_series = pd.read_csv('google_stock_price.csv',squeeze= True)
google_stock_price_series.head(6)
```

```python
def classified_performance(number):
    
    if number < 300:
        return "OK"
    elif number >= 300 and number < 650:
        return "Satisfactory"
    else:
        return "Incridible"
```

```python
google_stock_price_series.apply(classified_performance)
```

```python
google_stock_price_series.apply(classified_performance).head()
```

```python
google_stock_price_series.apply(classified_performance).tail()
```

```python
google_stock_price_series.head(6)
```
