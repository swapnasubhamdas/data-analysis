```python
#Filter a DataFrame Based on A Condition
#Filter with More than One Condition (AND - &)
import pandas as pd

employee = pd.read_csv("employee.csv",parse_dates= ["Date of Birth","Date of Joining"])
select = ["Emp ID","First Name","Last Name","Gender","Date of Birth","Date of Joining","Salary","SSN","Last % Hike",
          "County","City","State"]
emp = employee[select]
emp["Gender"] = emp["Gender"].astype("category")
emp["State"] = emp["State"].astype("category")
emp["Last % Hike"] = emp["Last % Hike"].astype("category")
```

```python
emp.head()
```

```python
mask1 = emp["Gender"] == "M"
```

```python
mask2 = emp["State"] == "NY"
```

```python
mask3 = emp["Last % Hike"] == "14%"
```

```python
emp[mask1 & mask2 & mask3]
```

```python
#Filter with More than One Condition (OR - |)
```

```python
emp[mask1 | mask2].head(10)
```

```python
emp[(mask1 & mask2) | mask3].head(16)
```

```python
#The .isin() Method
import pandas as pd

employee = pd.read_csv("employee.csv",parse_dates= ["Date of Birth","Date of Joining"])
select = ["Emp ID","First Name","Last Name","Gender","Date of Birth","Date of Joining","Salary","SSN","Last % Hike",
          "County","City","State"]
emp = employee[select]
emp["Gender"] = emp["Gender"].astype("category")
emp["State"] = emp["State"].astype("category")
emp["Last % Hike"] = emp["Last % Hike"].astype("category")
```

```python
e = emp.sort_values(by ="Emp ID",inplace= False)
```

```python
e.head()
```

```python
mask1 = e["State"] == "NY"
mask2 = e["State"] == "VA"
mask3 = e["State"] == "NH"
```

```python
e[mask1 | mask2 | mask3].head(10)
```

```python
mask = e["State"].isin(["NY","NH","VA"])
```

```python
e[mask].head(20)
```

```python
#The .duplicated() Method
#Call the .duplicated() method to create a Boolean Series and use it to extract rows that have duplicate values.
import pandas as pd

employee = pd.read_csv("employee.csv",parse_dates= ["Date of Birth","Date of Joining"])
select = ["Emp ID","First Name","Last Name","Gender","Date of Birth","Date of Joining","Salary","SSN","Last % Hike",
          "County","City","State"]
emp = employee[select]
emp["Gender"] = emp["Gender"].astype("category")
emp["State"] = emp["State"].astype("category")
emp["Last % Hike"] = emp["Last % Hike"].astype("category")
emp.sort_values(by = ["First Name"], inplace= True)
emp.head(15)
```

```python
emp[~emp["First Name"].duplicated(keep = False)]
```

```python
# The .drop_duplicates() Method
emp.head()
```

```python
len(emp.drop_duplicates())
```

```python
len(emp)
```
