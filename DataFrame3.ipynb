```python
#A Review of the .apply() Method on Single Columns
import pandas as pd
bond = pd.read_csv("james_bond.csv", index_col= "Film")
bond.sort_index(inplace= True)
bond.head()
```

```python
def convert_number_to_string_and_add_million(number):
    return str(number)+" Million"
```

```python
bond["Box Office"] = bond["Box Office"].apply(convert_number_to_string_and_add_million)
bond["Budget"] = bond["Budget"].apply(convert_number_to_string_and_add_million)
```

```python
bond.head()
```

```python
bond["Bond Actor Salary"] = bond["Bond Actor Salary"].apply(convert_number_to_string_and_add_million)
```

```python
#A Review of the .apply() Method on Single Columns
import pandas as pd
bond = pd.read_csv("james_bond.csv", index_col= "Film")
bond["Bond Actor Salary"] = bond["Bond Actor Salary"].fillna(value= "Unknown")
bond.sort_index(inplace= True)
bond.head()
```

```python
columns = ["Box Office","Budget","Bond Actor Salary"]
```

```python
for col in columns:
    bond[col] = bond[col].apply(convert_number_to_string_and_add_million)
```

```python
bond.head()
```

```python
#The .apply() Method with Row Values
import pandas as pd
bond = pd.read_csv("james_bond.csv", index_col= "Film")
bond.sort_index(inplace= True)
bond.head()
```

```python
def review(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The Best Bond Movie"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable Bond Movie"
    elif actor == "Sean Connery":
        return "Classic Bond Movie"
    else:
        return "I have no clue"
```

```python
bond["Review"] = bond.apply(review, axis="columns")
```

```python
bond.sort_values(by="Actor")
```
